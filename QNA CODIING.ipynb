**Q:2**
 Machine learning is a subset of artificial intelligence (AI) that involves the development of algorithms and statistical models that enable computers to learn from and make predictions

 Q:3**Supervised Learning**:    **HOW IT WORK**
The model is trained on a labeled dataset, which means that each training example is paired with an output label.
The goal is to learn a mapping from inputs to outputs that can be used to predict the labels of new, unseen data.



**UnSupervised Learning**: The model is trained on an unlabeled dataset, meaning the training data consists only of input variables without corresponding output labels.
The goal is to find hidden patterns or intrinsic structures in the input data.

Q:4**Supervised Learning**
1. Data Structure:

Uses labeled data, meaning each input comes with a corresponding output label.
2. Goal:

Learn a mapping from inputs to outputs that can be used to make predictions on new, unseen data.
3. Algorithms:

Regression: Linear regression, polynomial regression.
Classification: Logistic regression, decision trees, random forests, support vector machines (SVM), k-nearest neighbors (KNN), neural networks.
4. Applications:

Email spam detection: Classifying emails as spam or not spam.
Image and speech recognition: Identifying objects in images or transcribing spoken words into text.
Medical diagnosis: Predicting disease outcomes based on patient data.

**Unsupervised Learning**

The model is trained on an unlabeled dataset, meaning the training data consists only of input variables without corresponding output labels.
The goal is to find hidden patterns or intrinsic structures in the input data.
Common Algorithms:

Clustering: Groups data into clusters. Examples include k-means clustering, hierarchical clustering, and DBSCAN.
Association: Finds rules that describe large portions of data. An example is the Apriori algorithm.
Dimensionality Reduction: Reduces the number of features. Examples include principal component analysis (PCA) and t-SNE.
Applications:

Customer segmentation
Market basket analysis
Anomaly detection

 Q:5 **Regression**
1. Definition:

Regression is a type of supervised learning where the goal is to predict a continuous numeric value based on input features.
2. Goal:

To model the relationship between input features (independent variables) and a continuous output variable (dependent variable).
3. Output:

Continuous values, which can range over an infinite set of possible numbers.

**Classification**
1. Definition:

Classification is a type of supervised learning where the goal is to predict a categorical label based on input features.
2. Goal:

To model the relationship between input features and discrete output labels (classes).
3. Output:

Discrete labels, which are often categories or classes.

 Q5:**Training Set**
1. Definition:

The training set is a subset of the dataset used to train the machine learning model. It includes input features along with their corresponding output labels (in supervised learning).
2. Purpose:

To enable the model to learn the underlying patterns, relationships, and structures in the data.
The model adjusts its parameters based on this data to minimize prediction errors.
3. Use in Model Development:

The model is repeatedly exposed to the training set during the training phase.
The learning algorithm optimizes the model's parameters (e.g., weights in a neural network) to improve its predictions on the training data.
**Testing Set**
1. Definition:

The testing set is a subset of the dataset that is not used during the training phase. It includes input features and their corresponding output labels (in supervised learning).
2. Purpose:

To evaluate the performance and generalization capability of the trained model.
Provides an unbiased assessment of the model's ability to predict new, unseen data.
3. Use in Model Evaluation:

After the model is trained on the training set, it is tested on the testing set.
Performance metrics (e.g., accuracy, precision, recall, F1-score for classification; MSE, MAE, RÂ² for regression) are calculated to assess how well the model performs on data it has never seen before.
